3. Comments help make code readable. Use `#` for single-line comments and `''' '''` for multi-line.

# Lists:
A list is a collection which is ordered and mutable (changeable). Lists are created with square brackets `[]`.
You can store different types of data in a list.

### Methods: append(), remove(), pop(), sort(), reverse()

```python
fruits = ['Apple','Banana', 'Orange']
fruits
```

```python
fruits.append("Grapes")
fruits
```

```python
fruits.pop()
fruits
```

```python
fruits.sort()
fruits
```

```python
fruits.reverse()
fruits
```

```python
number = [1,22, 5, 67, 8]
number.sort()
number
```

# Tuples:
Tuples are immutable, meaning they can't be changed after creation. They use parentheses `()`.



```python
numbers = (7,14,21)
print(f"Unique numbers tuple: {numbers}")
```

# Sets:
A set is an unordered collection of unique elements. Created using curly braces `{}`.




```python
numbers = {1, 2, 3, 3, 4, 5}
print(f"Unique numbers set: {numbers}")
```

# Dictionaries:
Dictionaries store data as key-value pairs. Keys are unique, and values can be of any data type.


```python
student = {"name": "Alice", "age": 24, "course": "AI"}
student
```

# Functions:
Functions allow us to reuse code. We define a function using the `def` keyword.


```python
def greet(x):
  print("Hello "+x)


greet("Siri")
```

```python
def sum(x,y):
  print("Sum of X and Y is: ", x+y)

sum(3,8)
```

# Loops:
Loops let you iterate over data structures like lists, tuples, or dictionaries.

# Two common loops are `for` and `while`.

```python
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
  print(fruit)
```

```python
numbers = [ 4, 5, 6, 7, 8]
for num in numbers:
  print(num)
```

# Matplotlib Basics:
Matplotlib is used for creating graphs and visualizations.

```python
import matplotlib.pyplot as plt
```

```python
x = [ 12, 45, 32, 23, 56]
y = [10,20,30,40,50]
plt.plot(x,y)
plt.title("Line Plot")
plt.xlabel("X axis")
plt.ylabel(" Y axis")
plt.show()
```

```python
categories = ["A", "B", "C"]
values = [3, 7, 5]
plt.bar(categories, values)
plt.title("Bar Chart Example")
plt.show()
```
